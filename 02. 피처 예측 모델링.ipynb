# 피처 예측 모델링

```python
path='./data/'
data = pd.read_csv(path+'fin_data1.csv')
```

# ARIMA 함수 생성 및 적용

```python
from pmdarima.arima import auto_arima
def for_arima(dong, feature):
    df1 = data[['year','month','emd_nm',feature]]
#     i = df1[(df1['year']==2021)&(df1['month']==5)].index[0]
    
    train = df1.set_index(['year','month'])
    test = df1.set_index(['year','month'])
    
    train = train[train['emd_nm']==dong]
    train = train.drop(['emd_nm'],axis=1)
        
    model_arima= auto_arima(train,trace=True, error_action='ignore', start_p=1,start_q=1,max_p=3,max_q=3,
                            suppress_warnings=True,stepwise=False,seasonal=True)

    model = model_arima.fit(train[feature].values, trend='c', full_output=True, disp=True)
    fore = model.predict(steps=2)
    return fore[:2]
```

```python
cols = ['long_resd', 'long_work', 'long_visit','short_visit_pop_cnt','배달_cnt',
       '마트_cnt', '내국resd', '내국work', '내국visit', '거주foreign', '거주total','거주resid']
l1=[]
for i in data['emd_nm'].unique():
    for j in cols:
        l1.append(for_arima(i,j))
```

# 7,8월 DataFrame 생성

```python
#2021년 7월 dataframe 생성
a = pd.DataFrame(data['emd_nm'].unique())
a['year']=2021
a['month']=7
#2021년 8월 dataframe 생성
b = pd.DataFrame(data['emd_nm'].unique())
b['year']=2021
b['month']=8
```

```python
predict7 = pd.concat([a,pd.DataFrame(np.array(pd.DataFrame(l1)[0]).reshape(42,-1))],axis=1)
predict8 = pd.concat([b,pd.DataFrame(np.array(pd.DataFrame(l1)[1]).reshape(42,-1))],axis=1)
predict78 = pd.concat([predict7,predict8],axis=0)
predict78.columns = ['emd_nm','year','month','long_resd','long_work','long_visit','short_visit_pop_cnt',
                     '배달_cnt','마트_cnt','내국resd','내국work','내국visit','거주foreign','거주total','거주resid']
```

```python
공휴일 = pd.read_csv('/project/2021bigcontest/공휴일cnt.csv')
aa = pd.merge(predict78,공휴일,how='left',on=['year','month']).fillna(0)
aa = aa.rename(columns={'cnt':'holiday_cnt'})
aa['season'] = 1
```

# concat & save

```python
concat78 = pd.concat([data,aa],axis=0).sort_values(['year','month','emd_nm'])
concat78 = concat78[['year', 'month', 'emd_nm', 'long_resd', 'long_work', 'long_visit',
       'short_visit_pop_cnt', 'season', 'holiday_cnt', '배달_cnt',
       '마트_cnt', '내국resd', '내국work', '내국visit', '거주foreign', '거주total',
       '거주resid', 'em_g']]
concat78 = concat78.reset_index(drop=True)
# concat78.to_csv('/project/2021bigcontest/concat78.csv',index=False)
```
